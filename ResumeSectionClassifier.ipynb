```python
import pickle as pk
import numpy as np
import csv
import pandas as pd
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
```

```python
#I have taken only first 30 pickle files for training purpose and 5 pickle files for testing part.
```

```python
#Convert pickle files to csv

p=1
for i in range(1,31):
    with open("../input/data-resume/"+str(p)+".pickle","rb") as fp:
            A = pk.load(fp)
    Input = []
    Output = []

    for i in range(len(A[0])):
        Input.append(A[0][i])
        Output.append(A[1][i])

    A = np.array(Input)
    B = np.array(Output)
    C = np.reshape(B,(B.shape[0],1))
    D = np.concatenate((A,C),axis=1)
    E = np.asarray(D)
    Final = E.astype(np.object)

    df = pd.DataFrame(Final)
    df.to_csv("CSVFILE"+str(p)+".csv",index = False,header = None)
    p=p+1
```

```python
#Showing one of the csv file

show=pd.read_csv("CSVFILE1.csv",header=None)
show.head(5)
```

```python
# Feature selection part
# We have 16 features in total out of which we will select those which are relevant for our model
# Also, during training part we want numeric values so we will convert FALSE to 0 and TRUE to 1.
# Seeing csv files , i inferred that there was no significance of font style, font color and page ID
# When i trained model by including word position features, i was getting irrelevant output so i decided to drop those features too.
# So in next step we will drop the unwanted features
# Also we will drop wordID column because we don't need that during training part
```

```python
# Dropping unwanted columns

for i in range(1,31):
    y=pd.read_csv("CSVFILE"+str(i)+".csv",header = None)
    y=y.replace([False,True],[0,1])
    y=y.drop(y.columns[3], axis=1)
    y=y.drop(y.columns[3], axis=1)
    y=y.drop(y.columns[3], axis=1)
    y=y.drop(y.columns[0], axis=1)
    y=y.drop(y.columns[0], axis=1)
    y=y.drop(y.columns[0], axis=1)

    y.to_csv("CSVFILEUPDATE"+str(i)+".csv",index = False,header = None)
```

```python
# Showing one of he updated csv file

show=pd.read_csv("CSVFILEUPDATE1.csv",header=None)
show.head(5)
```

```python
#We have dropped 6 features and we are left with 10 features which we will use to train our model.
```

```python
#Merging all 30 csv files into one csv file

fout=open("MERGEDCSVFILE.csv","a")
for num in range(1,31):
    for line in open("CSVFILEUPDATE"+str(num)+".csv"):
         fout.write(line)    
fout.close()
```

```python
#Showing Merged CSV first 5 rows

Merge=pd.read_csv("MERGEDCSVFILE.csv",header=None)
Merge.head(5)
```

```python
#Taking all values of csv and splitting them into input X and output Y

DATASET = Merge.values
X_train = DATASET[:30521,:9]
Y_train = DATASET[:30521, 10]
```

```python
#Applying SVM through inbuilt library scikt learn

C = 1.0
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X_train, Y_train)
```

```python
#Testing part
# Converting pickle file to csv

p=31
for i in range(31,36):
    with open("../input/test-resume/"+str(p)+".pickle","rb") as fp:
            A = pk.load(fp)
    Input = []
    Output = []

    for i in range(len(A[0])):
        Input.append(A[0][i])
        Output.append(A[1][i])

    A = np.array(Input)
    B = np.array(Output)
    C = np.reshape(B,(B.shape[0],1))
    D = np.concatenate((A,C),axis=1)
    E = np.asarray(D)
    Final = E.astype(np.object)

    df = pd.DataFrame(Final)
    df.to_csv("FILETEST"+str(p)+".csv",index = False,header = None)
    p=p+1
```
